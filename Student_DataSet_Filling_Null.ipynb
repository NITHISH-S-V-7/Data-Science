```python
import numpy as np
import pandas as pd
```

FILLING NAN VALUES

```python
d={
    'sid':[101,102,103,104,105,106,107,108,109,110],
    'name':['madhumitha','vishnu','shan','Dharshini','sanjay','vikki','sathvik','mohan','narine','sumanth'],
    'phno':['9368564289','5387562748','8705926384','9074892547','7803685732','5384628564','6755364854',np.nan,np.nan,np.nan],
    'city':['chennai',np.nan,'hyd','kolkata',np.nan,'kashmir',np.nan,'bhglr','vizag','delhi'],
    'Gender':['Female','Female',np.nan,'Female','Male','Male',np.nan,np.nan,'Male','Male'],
    'domain':['py','java','py',np.nan,np.nan,np.nan,np.nan,'py','py','py']

}
df=pd.DataFrame(d)
df
```

```python
#Fill nan values in phno

df['phno']=df['phno'].fillna('Nat available')
df
```

```python
#filling nan values in city

df['city']=df['city'].fillna('unknown')
df
```

```python
#Filling nan values in Gender

#df['Gender']=df['Gender'].fillna('unknown')

df['Gender'].mode()

df['Gender']=df['Gender'].fillna(df['Gender'].mode()[0])
df
```

```python
#Filling nan values in donain
df['domain']=df['domain'].fillna(df['domain'].mode()[0])
df
```

FILLING NAN VALUES USING FORWARD AND BACKWARD FILL

```python
d={
    'sid':[101,102,103,104,105,106,107,108,109,110],
    'name':['madhumitha','vishnu','shan','Dharshini','sanjay','vikki','sathvik','mohan','narine','sumanth'],
    'city':['chennai',np.nan,'hyd','kolkata',np.nan,'kashmir',np.nan,'bhglr','vizag','delhi'],
    'Gender':['Female','Female',np.nan,'Female','Male','Male',np.nan,np.nan,'Male','Male'],
    'domain':['py','java','py',np.nan,np.nan,np.nan,np.nan,'py','py','py']

}
df=pd.DataFrame(d)
df
```

```python
#Filling nan values in city (Forwards)

df['city']=df['city'].ffill()
df
```

```python
#Filling nan values in Gender (Forward)

df['Gender']=df['Gender'].ffill()
df
```

```python
#Filling nan values in domain (Backward)

df['domain']=df['domain'].bfill()
df
```
